```python
from transformers import AutoTokenizer, AutoModel
import torch
import chromadb
import numpy as np

# تابع تولید Embedding
def get_embedding(text):
    tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
    model = AutoModel.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")

    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    embedding = outputs.last_hidden_state.mean(dim=1).numpy()

    # چک کردن نوع داده‌ها
    for value in embedding.flatten():
        if not isinstance(value, (int, float)):
            raise ValueError(f"Invalid type in embedding: {type(value)}")

    return embedding

# ایجاد ChromaDB client و collection
client = chromadb.Client()
collection = client.create_collection("texts")

# داده‌های نمونه
texts = ["Hello world", "This is a test"]

# تولید Embeddings
embeddings = [get_embedding(text) for text in texts]

# چک کردن Embeddings
for i, embedding in enumerate(embeddings):
    if not all(isinstance(value, (int, float)) for value in embedding.flatten()):
        raise ValueError(f"Invalid embedding at index {i}: {embedding}")

# اضافه کردن به ChromaDB
collection.add(
    embeddings=[embedding.tolist() for embedding in embeddings],
    metadatas=[{"text": text} for text in texts],
    ids=[str(i) for i in range(len(texts))]
)

print("Data added to ChromaDB successfully!")
```
